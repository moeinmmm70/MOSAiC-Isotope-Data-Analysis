```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Replace "your_file.xlsx" with the actual path to your Excel file
data = pd.read_excel("C:/CiASOM/PhD/07.Manuscripts/#2.Snow Surface/Data/Met City/RStudio/Hourly_Variables_2.xlsx")

# Update column labels
data = data.rename(columns={
    "bulk_Hl_10m_RunningHourlyMean": "Latent Heat Flux (W/m)",
    "temp_2m_RunningHourlyMean": "Air Temperature (C)",
    "rh_2m_RunningHourlyMean": "Relative Humidity (%)",
    "wspd_vec_mean_2m_RunningHourlyMean": "Wind Speed (m/s)",
    "snow_depth_RunningHourlyMean": "Snow Depth (cm)"
})

# Set the font to "Corbel"
font_name = "Corbel"

# Create subplots with shared x-axis (without the lowermost subplot)
fig, axes = plt.subplots(nrows=4, sharex=True, figsize=(10, 9))

# Plot each variable on its own subplot with different colors
latent_heat_flux = data["Latent Heat Flux (W/m)"]
latent_heat_flux_positive = latent_heat_flux.where(latent_heat_flux >= 0, 0)
latent_heat_flux_negative = latent_heat_flux.where(latent_heat_flux < 0, 0)

latent_heat_flux_positive_line, = axes[0].plot(data["DateTime"], latent_heat_flux_positive, color="#71B9B6", linewidth=0.5)
latent_heat_flux_negative_line, = axes[0].plot(data["DateTime"], latent_heat_flux_negative, color="#EE8CB3", linewidth=0.5)
axes[0].set_ylabel("Latent Heat Flux (W/m)", fontname=font_name, fontsize=12, color="#71B9B6")
axes[0].axhline(0, color="gray", linestyle="--", linewidth=0.5, alpha=0.3, zorder=0)

# Hide the fill between the lines
axes[0].fill_between(data["DateTime"], latent_heat_flux_positive, 0, color="none", alpha=0)
axes[0].fill_between(data["DateTime"], latent_heat_flux_negative, 0, color="none", alpha=0)

temperature_line, = axes[1].plot(data["DateTime"], data["Air Temperature (C)"], color="#BB3E03", linewidth=0.75)
axes[1].set_ylabel("Air Temperature (C)", fontname=font_name, fontsize=12, color=temperature_line.get_color())
axes[1].axhline(0, color="gray", linestyle="--", linewidth=0.5, alpha=0.3, zorder=0)
axes[1].axhline(-30, color="gray", linestyle="--", linewidth=0.5, alpha=0.3, zorder=0)

humidity_line, = axes[2].plot(data["DateTime"], data["Relative Humidity (%)"], color="#EF3C2D", linewidth=0.75)
axes[2].set_ylabel("Relative Humidity (%)", fontname=font_name, fontsize=12, color=humidity_line.get_color())
axes[2].axhline(95, color="gray", linestyle="--", linewidth=0.5, alpha=0.3, zorder=0)

# Plot Wind Speed 2m (m/s) with different colors based on the threshold
wind_speed = data["Wind Speed (m/s)"]
wind_speed_above_threshold = wind_speed.where(wind_speed >= 3, float("nan"))
wind_speed_below_threshold = wind_speed.where(wind_speed < 3, float("nan"))

axes[3].plot(data["DateTime"], wind_speed_above_threshold, color="#87153B", linewidth=0.75)
axes[3].plot(data["DateTime"], wind_speed_below_threshold, color="#9DCEE2", linewidth=0.75)
axes[3].set_ylabel("Wind Speed (m/s)", fontname=font_name, fontsize=12, color="#87153B")
axes[3].axhline(3.0, color="gray", linestyle="--", linewidth=0.5, alpha=0.3, zorder=0)

# Set common x-axis label
axes[-1].set_xlabel("Month", fontname=font_name, fontsize=20)

# Format the x-axis tick labels
date_format = mdates.DateFormatter("%b")
axes[-1].xaxis.set_major_formatter(date_format)

# Set ticks on every first day of the month
axes[-1].xaxis.set_major_locator(mdates.MonthLocator(bymonthday=1))

# Set the font for x-axis tick labels to "Corbel"
for label in axes[-1].xaxis.get_ticklabels():
    label.set_fontname('Corbel')

# Set the font for y-axis tick labels to "Corbel" for all subplots
for ax in axes:
    for label in ax.yaxis.get_ticklabels():
        label.set_fontname('Corbel')

# Add title
plt.suptitle("MET City", fontname=font_name, fontsize=20, fontweight='bold')

# Remove the box around each subplot
for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(ax.get_lines()[0].get_color())  # Set y-axis line color to match the line color
    if ax != axes[-1]:
        ax.spines['bottom'].set_visible(False)
        ax.tick_params(axis='x', which='both', bottom=False, top=False)  # Remove x-axis ticks

# Add grid lines at every second tick of x-axis
for ax in axes:
    ax.grid(axis='x', which='major', linestyle='--', linewidth=0.5, alpha=0.4)
    ax.set_axisbelow(True)

# Adjust spacing between subplots
plt.tight_layout()

# Set the font size and font family for tick labels on the y-axis
for ax in axes:
    ax.tick_params(axis='y', labelsize=12, colors=ax.get_lines()[0].get_color())

# Set the font size and font family for tick labels on the x-axis
for ax in axes:
    ax.tick_params(axis='x', labelsize=15, labelrotation=45)

# Save the plot as a JPEG file
plt.savefig("C:/CiASOM\PhD/07.Manuscripts/#2.Snow Surface/Figures/MET/TS_Plot_24012024.jpeg", dpi=600, bbox_inches='tight')

# Display the plot
plt.show()
```
