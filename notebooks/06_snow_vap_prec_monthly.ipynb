```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# File path and sheet name
file_path = r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Second Submission - Elementa\Revisions\Data_Figures\Surface_Snow_by_month.xlsx'
sheet_name = 'PREC_DIF_SNOW_VAP'

# Read the data
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract the necessary columns
x = df['prec_agg_month']
y = df['d18O']

# Plot
plt.figure(figsize=(10, 10))
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 18
plt.scatter(x, y, color='blue', alpha=0.7, edgecolor='black', s=100)

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept
plt.plot(x, line, color='red', linewidth=2, label=f'Linear Fit: O = {slope:.2f} X + {intercept:.2f}\n$R^2$ = {r_value**2:.2f}')

# Labels and title
plt.xlabel('X: Aggregated Monthly Precipitation Amount (mm)')
plt.ylabel('O ()')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

output_path = r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Second Submission - Elementa\Revisions\Data_Figures\delta_d18O_prec_amount.jpeg'
plt.savefig(output_path, dpi=600, format='jpeg')
plt.show()
```
