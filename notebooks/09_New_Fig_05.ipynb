```python
import os
from PIL import Image, ImageDraw, ImageFont
from fpdf import FPDF

# File paths
folder_path = r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Second Submission - Elementa\Revisions\Data_Figures'
file_names = ['fig_05_ precip_amount_TS.jpg', 'fig_05_prec_snow_vap.jpg']
output_pdf_path = os.path.join(folder_path, 'Combined_Figures.pdf')

# Load images
images = [Image.open(os.path.join(folder_path, file_name)) for file_name in file_names]

# Define font
font_path = 'arial.ttf'  # Adjust the path if necessary
font_size = 220
font = ImageFont.truetype(font_path, font_size)

# Add labels to images
labels = ['(a)', '(b)']
for img, label in zip(images, labels):
    draw = ImageDraw.Draw(img)
    draw.text((20, 50), label, font=font, fill='black')

# Create a new blank image with the width of two images and the height of the tallest
new_width = sum(img.width for img in images)
new_height = max(img.height for img in images)
new_image = Image.new('RGB', (new_width, new_height), 'white')

# Paste images next to each other
x_offset = 0
for img in images:
    new_image.paste(img, (x_offset, 0))
    x_offset += img.width

# Save as a PDF
new_image.save(output_pdf_path, 'PDF')

print(f'PDF created at {output_pdf_path}')
```
