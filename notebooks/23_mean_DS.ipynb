```python
import pandas as pd
import os

# File paths
excel_file_path = r"C:\CiASOM\PhD\07.Manuscripts\#1.Overview\Figures\06_Figure06_seaice\DS_FYI_SYI.xlsx"
csv_output_path = r"C:\CiASOM\PhD\07.Manuscripts\#1.Overview\Figures\06_Figure06_seaice\Mean_DS_SYI.csv"

# Read the Excel file
sheet_name = "DS_SYI"
df = pd.read_excel(excel_file_path, sheet_name)

# Calculate mean, standard deviation, min, and max values in 10 cm intervals
interval = 10  # cm
min_depth = df["Mean_Depth"].min()
max_depth = df["Mean_Depth"].max()
depth_intervals = range(int(min_depth), int(max_depth) + interval, interval)

result_data = []
for start_depth in depth_intervals:
    end_depth = start_depth + interval
    interval_data = df[(df["Mean_Depth"] >= start_depth) & (df["Mean_Depth"] < end_depth)]
    if not interval_data.empty:
        mean_d18O = interval_data["d18O"].mean()
        std_d18O = interval_data["d18O"].std()
        min_d18O = interval_data["d18O"].min()
        max_d18O = interval_data["d18O"].max()

        mean_d_excess = interval_data["d_excess"].mean()
        std_d_excess = interval_data["d_excess"].std()
        min_d_excess = interval_data["d_excess"].min()
        max_d_excess = interval_data["d_excess"].max()

        mean_salinity = interval_data["Salinity (psu)"].mean()
        std_salinity = interval_data["Salinity (psu)"].std()
        min_salinity = interval_data["Salinity (psu)"].min()
        max_salinity = interval_data["Salinity (psu)"].max()

        result_data.append([
            start_depth, end_depth,
            mean_d18O, std_d18O, min_d18O, max_d18O,
            mean_d_excess, std_d_excess, min_d_excess, max_d_excess,
            mean_salinity, std_salinity, min_salinity, max_salinity
        ])

# Create a DataFrame for the results
result_columns = [
    "Start_Depth", "End_Depth",
    "Mean_d18O", "Std_d18O", "Min_d18O", "Max_d18O",
    "Mean_d_excess", "Std_d_excess", "Min_d_excess", "Max_d_excess",
    "Mean_Salinity", "Std_Salinity", "Min_Salinity", "Max_Salinity"
]
result_df = pd.DataFrame(result_data, columns=result_columns)

# Save results to a CSV file
result_df.to_csv(csv_output_path, index=False)

print("Results saved to", csv_output_path)
```
