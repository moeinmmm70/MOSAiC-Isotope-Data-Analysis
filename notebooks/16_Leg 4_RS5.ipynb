```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
import matplotlib.dates as mdates
import matplotlib as mpl

# Set the global font to Corbel
mpl.rcParams['font.family'] = 'Corbel'

# Read the Excel file
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected.xlsx"
df = pd.read_excel(file_path, sheet_name='All')

# Filter the data for specific locations
locations = ["RS5-A1", "RS5-A2", "RS5-A3", "RS5-A4"]
filtered_df = df[df['Location'].isin(locations)].copy()

# Convert 'Date' to datetime
filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])

# Calculate the mean of d18O values
mean_d18O = filtered_df['d18O'].mean()

# Calculate the anomalies (difference from the mean)
filtered_df['d18O_anomaly'] = filtered_df['d18O'] - mean_d18O

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plot using the axes object, disable the legend, and set edgecolor to black
sns.scatterplot(data=filtered_df, x='Date', y='Mean_Depth', hue='d18O_anomaly', palette='rainbow', ax=ax, 
                legend=False, edgecolor='black')

# Create a colorbar
norm = mcolors.Normalize(vmin=filtered_df['d18O_anomaly'].min(), vmax=filtered_df['d18O_anomaly'].max())
sm = plt.cm.ScalarMappable(cmap='rainbow', norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, label="$^{18}$O () (Anomaly From Mean)")

# Set x-axis to weekly intervals
ax.xaxis.set_major_locator(mdates.WeekdayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Add gridlines with specified style
ax.grid(True, linestyle='--', linewidth=0.3, color='#CCCCCC')  # 20% grey color

# Aesthetics
ax.set_title('Leg 4: RS5 Stations')
ax.set_xlabel('Date')
ax.set_ylabel('Mean Height (cm)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
import matplotlib.dates as mdates
import matplotlib as mpl

# Set the global font to Corbel
mpl.rcParams['font.family'] = 'Corbel'

# Read the first Excel file (daily data)
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected.xlsx"
df = pd.read_excel(file_path, sheet_name='All')

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame for the specified date range
start_date = '2020-06-17'
end_date = '2020-07-31'
date_filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()

# Read the second Excel file (hourly data)
file_path2 = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Met City\RStudio\Hourly_Variables_2.xlsx"
df2 = pd.read_excel(file_path2)

# Convert 'DateTime' to datetime for the second DataFrame and extract date
df2['DateTime'] = pd.to_datetime(df2['DateTime'], format='%Y-%m-%d %H:%M:%S')
df2['Date'] = df2['DateTime'].dt.date  # Extract just the date part for daily aggregation

# Calculate daily means for temp_2m_RunningHourlyMean
daily_means = df2.groupby('Date')['temp_2m_RunningHourlyMean'].mean().reset_index()
daily_means['Date'] = pd.to_datetime(daily_means['Date'])  # Convert back to datetime for merging

# Merge the two DataFrames on 'Date'
merged_df = pd.merge(date_filtered_df, daily_means, on='Date', how='left')

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the first dataset using the axes object
sns.scatterplot(data=merged_df, x='Date', y='Mean_Depth', hue='d18O', palette='GnBu', ax=ax, 
                legend=False, edgecolor='black')

# Create a colorbar for d18O and adjust its position
norm = mcolors.Normalize(vmin=merged_df['d18O'].min(), vmax=merged_df['d18O'].max())
sm = plt.cm.ScalarMappable(cmap='GnBu', norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, label="$^{18}$O ()")
cbar.ax.set_position([0.82, 0.15, 0.03, 0.7])  # Adjust position [left, bottom, width, height]

# Set up the primary y-axis for Mean Depth
ax.set_ylabel('Mean Snow Height (cm)')

# Set up the secondary y-axis for Mean Daily Air Temperature
ax2 = ax.twinx()
ax2.plot(merged_df['Date'], merged_df['temp_2m_RunningHourlyMean'], color='#F26A4F', alpha=0.5)
ax2.set_ylabel('Mean Daily Air Temperature 2m (C)', color='#F26A4F')  # Set y-axis label color
ax2.tick_params(axis='y', colors='#F26A4F')  # Set tick color to red
ax2.spines['right'].set_color('#F26A4F')  # Change the color of the right spine color

# Add gridlines and other aesthetics
ax.grid(True, linestyle='--', linewidth=0.3, color='#CCCCCC')
ax.set_title('Leg 4')
ax.set_xlabel('Date')

# Format the x-axis to show dates
ax.xaxis.set_major_locator(mdates.WeekdayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Save the plot as a JPEG file
plt.savefig(r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Figures\Leg4_Leg5/Leg_4.jpeg", dpi=600, bbox_inches='tight')

plt.show()
```
