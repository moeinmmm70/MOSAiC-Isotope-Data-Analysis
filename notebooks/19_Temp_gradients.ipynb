```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime


fig, ax1 = plt.subplots(figsize=(11, 5), dpi=300)

fnameT = 'Y:\mosaic\data\metadata\Final\metadata_Temperature.xlsx'
temp_df = pd.read_excel(fnameT,'MOSAiC_ICE_Temperature')

fname = 'Y:\mosaic\data\processed\snowpits\isotopes\MacMellat\Data\ALL_Isotopes.xlsx'
iso_df = pd.read_excel(fname,'Sheet2')

continueDF = pd.DataFrame()

temp_df['AtSnowHeight cm']=temp_df['AtSnowHeight cm'].astype(float)
temp_df=temp_df.dropna(subset=['AtSnowHeight cm', 'Temperature degC', 'Timestamp'])
temp_df = temp_df[temp_df['Timestamp'] < '2020-05-15'] 

unique_device_ids = pd.unique(temp_df['Event'])
for udi in unique_device_ids:
    #try:
        temp_df2=temp_df[temp_df['Event']== udi]
        temp_df2=temp_df2.sort_values(by=['AtSnowHeight cm'], ascending=False)
        temp_df2['Timestamp'] = temp_df2['Timestamp'].astype('datetime64[ns]')

        h_max, t_max = temp_df2['AtSnowHeight cm'].iloc[0], temp_df2['Temperature degC'].iloc[0]
        h_min, t_min = temp_df2['AtSnowHeight cm'].iloc[-1], temp_df2['Temperature degC'].iloc[-1]
        temp_diff = t_max-t_min
        h_diff = h_max-h_min
        temp_grad = temp_diff/(h_diff/100)
        if temp_grad > 0:
            print(udi, temp_grad)
        if h_diff > 1:
            plt.plot([temp_df2['Timestamp'].iloc[0],temp_df2['Timestamp'].iloc[0]], [0, temp_grad], linestyle='--', c='lightgrey')
            plt.scatter(temp_df2['Timestamp'].iloc[0], temp_grad, c=h_diff,marker='x', cmap='PiYG',vmin = 5, vmax=18)
        else:
            continue
        
  #  except:
   #     continue
        
plt.colorbar(label='Snowpit height (cm)')
plt.xticks(rotation=30)
plt.xlabel('Datetime (YYYY-MM)')
plt.ylabel('Temperature gradient (K/m)')
plt.axhline(0, c='grey', linestyle='--')
```
