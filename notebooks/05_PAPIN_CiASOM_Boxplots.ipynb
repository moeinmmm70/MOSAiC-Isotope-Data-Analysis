```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the Excel file
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Precipitation_PAPIN\CiASOM_Comparison/PAPIN_Surface_CiASOM_1.xlsx"
df = pd.read_excel(file_path, sheet_name="Pallas")

# Convert the "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract month from the date
df['Month'] = df['Date'].dt.to_period('M')  # Use periods to represent months

# Create a custom order for months
custom_order = pd.period_range('2019-10', '2020-09', freq='M')

# Calculate mean for each month
mean_values = [df[df['Month'] == i]['d18O'].mean() for i in custom_order]

# Create a boxplot with blue filling
plt.figure(figsize=(10, 6))
box = plt.boxplot([df[df['Month'] == i]['d18O'] for i in custom_order],
                  labels=[month.strftime('%B %Y') for month in custom_order], patch_artist=True, showfliers=False)

# Set the box color to blue
for patch in box['boxes']:
    patch.set_facecolor('blue')

# Change the thickness of the boxes to 2.0
for box_elem in box['boxes']:
    box_elem.set_linewidth(2.0)

# Change the color of the box lines to gray (70%)
for box_elem in box['boxes']:
    box_elem.set_edgecolor((0.7, 0.7, 0.7))

# Change the thickness of the whiskers to 2.0
for whisker_elem in box['whiskers']:
    whisker_elem.set_linewidth(2.0)

# Change the color of the whiskers to gray (70%)
for whisker_elem in box['whiskers']:
    whisker_elem.set_color((0.7, 0.7, 0.7))

# Change title, labels, and fonts
plt.title('Monthly Distribution of $\\delta^{18}O$', fontname='Corbel', fontsize=16, fontstyle='italic')
plt.xlabel('Month', fontname='Corbel', fontsize=14)
plt.ylabel('$\\delta^{18}O$ (%)', fontname='Corbel', fontsize=14)
plt.xticks(rotation=45, fontname='Corbel', fontsize=12)
plt.yticks(fontname='Corbel', fontsize=12)

# Add plus sign markers for outliers
for flier in box['fliers']:
    flier.set(marker='+', markersize=8, markeredgecolor='blue')

# Change the middle line to mean
for median in box['medians']:
    median.set(color='cyan', linewidth=2.0, linestyle='dashed')

# Change the caps (whiskers) to mean
for cap in box['caps']:
    cap.set(color='cyan', linewidth=2.0)

# Add grid lines
plt.grid(axis='x', linestyle='dashed', linewidth=0.8, alpha=0.3)
plt.grid(axis='y', linestyle='dotted', linewidth=0.5, alpha=0.2)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.lines as mlines

def create_boxplot(ax, dfs, column_name, custom_order, x_positions, colors, width=0.3):
    for i, (df, color) in enumerate(zip(dfs, colors)):
        box = ax.boxplot([df[df['Month'] == i][column_name] for i in custom_order],
                         positions=x_positions,
                         patch_artist=True, showfliers=False, widths=width)

        for box_elem in box['boxes']:
            box_elem.set(color=color, linewidth=1.5, alpha=0.7)
            box_elem.set_facecolor(color)

        for whisker_elem in box['whiskers']:
            whisker_elem.set(color=color, linewidth=1.5, alpha=0.7)

        for median in box['medians']:
            median.set(color=color, linewidth=1.5)

        for cap in box['caps']:
            cap.set(color=color, linewidth=1.5, alpha=0.7)

    ax.set_xticks(x_positions)
    ax.set_xticklabels([month.strftime('%b') for month in custom_order])

    return ax

# Read the Excel file
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Precipitation_PAPIN\CiASOM_Comparison\PAPIN_Surface_CiASOM_3.xlsx"
df_pallas = pd.read_excel(file_path, sheet_name="Pallas")
df_ny_alesund = pd.read_excel(file_path, sheet_name="Ny-lesund")
df_samoylov = pd.read_excel(file_path, sheet_name="Samoylov")
df_surface_CiASOM = pd.read_excel(file_path, sheet_name="Surface_CiASOM")

# Convert the "Date" column to datetime format and extract month
dfs = [df_pallas, df_ny_alesund, df_samoylov, df_surface_CiASOM]
for df in dfs:
    df['Date'] = pd.to_datetime(df['Date'])
    df['Month'] = df['Date'].dt.to_period('M')

# Create a custom order for months
custom_order = pd.period_range('2019-10', '2020-09', freq='M')

# Define x-axis positions
x_positions = np.arange(len(custom_order))

# Start of subplot creation with minimal spacing between subplots
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(12, 10))
fig.subplots_adjust(hspace=0.05)  # Adjust the horizontal space between subplots

# Colors for each dataset
colors = ["#CB1B16", "#1368AA", "#94d2bd"]

# Subplot for d18O
axs[0] = create_boxplot(axs[0], [df_samoylov, df_ny_alesund, df_surface_CiASOM], 'd18O', custom_order, x_positions, colors)
axs[0].set_ylabel('$\\delta^{18}O$ ()', fontname='Corbel', fontsize=14)

# Create legend entries
legend_entries = [
    mlines.Line2D([0], [0], color="#CB1B16", linewidth=1.5, alpha=0.7, label='Samoylov Island (N: 90)'),
    mlines.Line2D([0], [0], color="#1368AA", linewidth=1.5, alpha=0.7, label='Ny-lesund (N: 25)'),
    mlines.Line2D([0], [0], color="#94d2bd", linewidth=1.5, alpha=0.7, label='Surface Snow MOSAiC (N: 213)')
]

# Add legend inside the upper subplot
axs[0].legend(handles=legend_entries, loc='upper left')

# Subplot for d-excess
axs[1] = create_boxplot(axs[1], [df_samoylov, df_ny_alesund, df_surface_CiASOM], 'd-excess', custom_order, x_positions, colors)
axs[1].set_xlabel('Month', fontname='Corbel', fontsize=14)
axs[1].set_ylabel('d-excess ()', fontname='Corbel', fontsize=14)

# Set x-ticks and rotation
plt.xticks(rotation=45, fontname='Corbel', fontsize=12)

# Add grid lines
for ax in axs:
    ax.grid(axis='y', linestyle='dashed', linewidth=0.5, alpha=0.2)
    
# Add labels "a)" and "b)" next to the subplots
fig.text(0.072, 0.865, "(a)", fontdict={'fontname': 'Times New Roman', 'fontsize': 18}, transform=fig.transFigure)
fig.text(0.072, 0.465, "(b)", fontdict={'fontname': 'Times New Roman', 'fontsize': 18}, transform=fig.transFigure)

# Save the plot as a JPEG with 600 dpi
output_file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Precipitation_PAPIN\CiASOM_Comparison\M_Box_CiA_Sam_Ny_d18O_dex_04032024.jpg"
plt.savefig(output_file_path, dpi=600, bbox_inches='tight')
print(f"Plot saved at: {output_file_path}")

# Show the plot
plt.show()
```

```python
import pandas as pd

# Read the Excel file
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Precipitation_PAPIN\CiASOM_Comparison\PAPIN_Surface_CiASOM_3.xlsx"
df_ny_alesund = pd.read_excel(file_path, sheet_name="Ny-lesund")
df_samoylov = pd.read_excel(file_path, sheet_name="Samoylov")
df_surface_CiASOM = pd.read_excel(file_path, sheet_name="Surface_CiASOM")

# Convert the "Date" column to datetime format and extract month
dfs = [df_ny_alesund, df_samoylov, df_surface_CiASOM]
for df in dfs:
    df['Date'] = pd.to_datetime(df['Date'])
    df['Month'] = df['Date'].dt.to_period('M')

# Create a custom order for months
custom_order = pd.period_range('2019-10', '2020-09', freq='M')

# Dataframes and their names
dfs = [df_ny_alesund, df_samoylov, df_surface_CiASOM]
df_names = ['Ny-lesund', 'Samoylov', 'Surface_CiASOM']

# Function to calculate median, min, and max for a given column in a dataframe
def calculate_monthly_stats(df, column_name):
    grouped = df.groupby('Month')[column_name]
    return grouped.median(), grouped.min(), grouped.max()

# Calculate and store median, min, and max values for each dataframe and for each parameter
stats = {'Median': {}, 'Min': {}, 'Max': {}}
for column_name in ['d18O', 'd-excess']:
    for df, df_name in zip(dfs, df_names):
        median, min_val, max_val = calculate_monthly_stats(df, column_name)
        stats['Median'][(column_name, df_name)] = median
        stats['Min'][(column_name, df_name)] = min_val
        stats['Max'][(column_name, df_name)] = max_val

# Convert each inner dictionary into a DataFrame and then concatenate
median_df = pd.concat({k: pd.DataFrame(v) for k, v in stats['Median'].items()}, axis=1)
min_df = pd.concat({k: pd.DataFrame(v) for k, v in stats['Min'].items()}, axis=1)
max_df = pd.concat({k: pd.DataFrame(v) for k, v in stats['Max'].items()}, axis=1)

# Concatenate median, min, and max DataFrames
combined_stats_df = pd.concat([median_df, min_df, max_df], axis=1, keys=['Median', 'Min', 'Max'])

# Print the combined stats dataframe
print(combined_stats_df)

# Save the combined stats dataframe to an Excel file
output_file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Precipitation_PAPIN\CiASOM_Comparison\combined_stats.xlsx"  # Replace with your desired file path
combined_stats_df.to_excel(output_file_path, sheet_name='Monthly Stats')

print(f"Data saved to {output_file_path}")
```
