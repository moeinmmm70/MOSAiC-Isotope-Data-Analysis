```python
import os
import pandas as pd

# Directory path containing the CSV files
directory = r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Device_Op_Num_files'

# List all CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Initialize an empty DataFrame to store the selected rows
selected_rows = pd.DataFrame()

# Iterate over each CSV file
for file in csv_files:
    file_path = os.path.join(directory, file)
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Find the row(s) with the highest "Mean_Depth" value
    max_mean_depth = df['Mean_Depth'].max()
    max_rows = df[df['Mean_Depth'] == max_mean_depth]
    
    # Append the selected rows to the DataFrame
    selected_rows = pd.concat([selected_rows, max_rows])

# Output file path for the selected rows
output_file = r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Surface_Snow_Samples.csv'

# Write the selected rows to a new CSV file
selected_rows.to_csv(output_file, index=False)

print(f"Selected rows from {len(csv_files)} CSV files have been saved to {output_file}.")
```

```python
import pandas as pd

# Read the data from Excel file
data = pd.read_excel(r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected.xlsx',
                     sheet_name='Surface_Snow')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# Select the columns for mean calculation
columns_to_mean = ['Lat', 'Lon', 'Air_Temp', 'd18O', 'dD', 'd_excess']

# Resample the data to weekly frequency starting from Sunday and calculate the means
weekly_data = data[columns_to_mean].resample('W-SUN').mean().reset_index()

# Save the data as a CSV file
weekly_data.to_csv(r'C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\weekly_surface_snow.csv', index=False)
```
