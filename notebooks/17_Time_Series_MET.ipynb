```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
data = pd.read_excel("C:/CiASOM/PhD/07.Manuscripts/#2.Snow Surface/Data/Met City/RStudio/Hourly_Variables_2.xlsx")

# Convert the data to long format
data_long = pd.melt(data,
                    value_vars=["bulk_Hl_10m_RunningHourlyMean",
                                "rh_2m_RunningHourlyMean",
                                "snow_depth_RunningHourlyMean",
                                "temp_2m_RunningHourlyMean",
                                "wspd_vec_mean_2m_RunningHourlyMean"],
                    var_name="Variable",
                    value_name="Value")

# Update variable labels
data_long["Variable"] = data_long["Variable"].replace({
    "bulk_Hl_10m_RunningHourlyMean": "Latent Heat Flux (W/m2)",
    "rh_2m_RunningHourlyMean": "Relative Humidity 2m (%)",
    "snow_depth_RunningHourlyMean": "Snow Depth (cm)",
    "temp_2m_RunningHourlyMean": "Air Temperature 2m (°C)",
    "wspd_vec_mean_2m_RunningHourlyMean": "Wind Speed 2m (m/s)"
})

# Create a time series plot using matplotlib
fig, ax = plt.subplots()
for variable, subset in data_long.groupby("Variable"):
    ax.plot(subset["DateTime"], subset["Value"], label=variable)
ax.set_xlabel("Date/Time")
ax.set_ylabel("Value")
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Replace "your_file.xlsx" with the actual path to your Excel file
data = pd.read_excel("C:/CiASOM/PhD/07.Manuscripts/#2.Snow Surface/Data/Met City/RStudio/Hourly_Variables_2.xlsx")

# Update column labels
data = data.rename(columns={
    "bulk_Hl_10m_RunningHourlyMean": "Latent Heat Flux (W/m²)",
    "temp_2m_RunningHourlyMean": "Air Temperature 2m (°C)",
    "rh_2m_RunningHourlyMean": "Relative Humidity 2m (%)",
    "wspd_vec_mean_2m_RunningHourlyMean": "Wind Speed 2m (m/s)",
    "snow_depth_RunningHourlyMean": "Snow Depth (cm)"
})

# Set the font to "Cambria"
font_name = "Cambria"

# Create subplots with shared x-axis
fig, axes = plt.subplots(nrows=5, sharex=True, figsize=(10, 12))

# Plot each variable on its own subplot with different colors
latent_heat_flux = data["Latent Heat Flux (W/m²)"]
latent_heat_flux_positive = latent_heat_flux.where(latent_heat_flux >= 0, 0)
latent_heat_flux_negative = latent_heat_flux.where(latent_heat_flux < 0, 0)

latent_heat_flux_positive_line, = axes[0].plot(data["DateTime"], latent_heat_flux_positive, color="darkblue", linewidth=0.5)
latent_heat_flux_negative_line, = axes[0].plot(data["DateTime"], latent_heat_flux_negative, color="pink", linewidth=0.5)
axes[0].set_ylabel("Latent Heat Flux (W/m²)", fontname=font_name, fontsize=12, color="darkblue")

# Hide the fill between the lines
axes[0].fill_between(data["DateTime"], latent_heat_flux_positive, 0, color="none", alpha=0)
axes[0].fill_between(data["DateTime"], latent_heat_flux_negative, 0, color="none", alpha=0)

temperature_line, = axes[1].plot(data["DateTime"], data["Air Temperature 2m (°C)"], color="purple")
axes[1].set_ylabel("Air Temperature 2m (°C)", fontname=font_name, fontsize=12, color=temperature_line.get_color())
axes[1].axhline(0, color="gray", linestyle="--", linewidth=0.75, alpha=0.3, zorder=0)
axes[1].axhline(-30, color="gray", linestyle="--", linewidth=0.75, alpha=0.3, zorder=0)

humidity_line, = axes[2].plot(data["DateTime"], data["Relative Humidity 2m (%)"], color="orange")
axes[2].set_ylabel("Relative Humidity 2m (%)", fontname=font_name, fontsize=12, color=humidity_line.get_color())
axes[2].axhline(95, color="gray", linestyle="--", linewidth=0.75, alpha=0.3, zorder=0)

wind_speed_line, = axes[3].plot(data["DateTime"], data["Wind Speed 2m (m/s)"], color="firebrick")
axes[3].set_ylabel("Wind Speed 2m (m/s)", fontname=font_name, fontsize=12, color=wind_speed_line.get_color())
axes[3].axhline(5.0, color="gray", linestyle="--", linewidth=0.75, alpha=0.3, zorder=0)

snow_depth_line, = axes[4].plot(data["DateTime"], data["Snow Depth (cm)"], color="lightblue")
axes[4].set_ylabel("Snow Depth (cm)", fontname=font_name, fontsize=12, color=snow_depth_line.get_color())
axes[4].axhline(0, color="gray", linestyle="--", linewidth=0.75, alpha=0.3, zorder=0)

# Set common x-axis label
axes[-1].set_xlabel("Date", fontname=font_name, fontsize=20, fontweight='bold')

# Format the x-axis tick labels
date_format = mdates.DateFormatter("%d/%m/%Y")
axes[-1].xaxis.set_major_formatter(date_format)

# Set ticks on every first day of the month
axes[-1].xaxis.set_major_locator(mdates.MonthLocator(bymonthday=1))

# Specify the desired start and end dates
start_date = pd.to_datetime("2019-10-01")
end_date = pd.to_datetime("2020-06-01")

# Update the x-axis limits for all subplots
for ax in axes:
    ax.set_xlim(start_date, end_date)

# Add title
plt.suptitle("MET City", fontname=font_name, fontsize=20, fontweight='bold')

# Remove the box around each subplot
for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(ax.get_lines()[0].get_color())  # Set y-axis line color to match the line color
    if ax != axes[-1]:
        ax.spines['bottom'].set_visible(False)
        ax.tick_params(axis='x', which='both', bottom=False, top=False)  # Remove x-axis ticks

# Add grid lines at every second tick of x-axis
for ax in axes:
    ax.grid(axis='x', which='major', linestyle='--', linewidth=0.5, alpha=0.4)
    ax.set_axisbelow(True)
    #ax.xaxis.set_major_locator(mdates.DayLocator(interval=60))

# Adjust spacing between subplots
plt.tight_layout()

# Set the font size and font family for tick labels on the y-axis
for ax in axes:
    ax.tick_params(axis='y', labelsize=12, colors=ax.get_lines()[0].get_color())

# Set the font size and font family for tick labels on the x-axis
for ax in axes:
    ax.tick_params(axis='x', labelsize=15, labelrotation=45)

# Save the plot as a JPEG file
plt.savefig("C:/CiASOM/PhD/07.Manuscripts/#2.Snow Surface/Data/Met City/RStudio/TS_Plot_Leg123.jpeg", dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Met City\RStudio\Hourly_MET.xlsx"
your_data = pd.read_excel(file_path)
your_data['DateTime'] = pd.to_datetime(your_data['DateTime'])
your_data['month'] = your_data['DateTime'].dt.month_name()

all_months = pd.DataFrame({'month': pd.date_range(start='2019-09-01', end='2020-10-01', freq='MS').month_name()})
custom_order = pd.period_range('2019-09', '2020-08', freq='M')
x_positions = np.arange(len(custom_order))

your_data = pd.merge(all_months, your_data, on='month', how='left')

variables = ["temp_2m_RunningHourlyMean", "rh_2m_RunningHourlyMean", "wspd_vec_mean_2m_RunningHourlyMean"]
labels = {"temp_2m_RunningHourlyMean": "Air Temperature 2m (°C)",
          "rh_2m_RunningHourlyMean": "Relative Humidity 2m (%)",
          "wspd_vec_mean_2m_RunningHourlyMean": "Wind Speed 2m (m/s)"}
variable_colors = {"temp_2m_RunningHourlyMean": "#FE2C01",
                   "rh_2m_RunningHourlyMean": "#ECD3AD",
                   "wspd_vec_mean_2m_RunningHourlyMean": "#022A5D"}

fig, axes = plt.subplots(nrows=len(variables), ncols=1, figsize=(10, 10), sharex=True)

for i, variable in enumerate(variables):
    ax = axes[i]
    monthly_data = [your_data[variable][your_data['month'] == month].dropna() for month in your_data['month'].unique()]
    color = variable_colors[variable]
    
    violin = ax.violinplot(monthly_data, positions=np.arange(1, len(custom_order)+1),
                           showmeans=False, showmedians=True, showextrema=True)
    
    for pc in violin['bodies']:
        pc.set_facecolor(color)
        pc.set_edgecolor('#464C66')
        pc.set_alpha(0.7)

    # Calculate and plot outliers
    for j, month_data in enumerate(monthly_data):
        q1 = np.percentile(month_data, 25)
        q3 = np.percentile(month_data, 75)
        iqr = q3 - q1
        outliers = month_data[(month_data < q1 - 1.5 * iqr) | (month_data > q3 + 1.5 * iqr)]
        ax.scatter([j + 1] * len(outliers), outliers, marker='o', color='grey')

    ax.set_ylabel(labels[variable], fontdict={'family': 'Corbel', 'rotation': 90, 'fontsize': 15.0})
    ax.yaxis.grid(True, linestyle='--', alpha=0.2, color='grey')
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_xticks(np.arange(1, len(custom_order)+1))
    ax.set_xticklabels([month.strftime('%B %Y') for month in custom_order], family='Corbel', rotation=0, size=12)

axes[-1].set_xlabel('Month', fontdict={'family': 'Corbel', 'fontsize': 15.0})
plt.xticks(rotation=90, fontname='Corbel', fontsize=12)
plt.tight_layout()

output_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Figures\MET\MET_violin_monthly_1.jpg"
plt.savefig(output_path, dpi=600)
plt.show()
```
