```python
import pandas as pd

# Read the Excel file
file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected.xlsx"
sheet_name = "All"
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Group the data by "PS Operation Device/Lable on Bottle"
groups = df.groupby("PS Operation Device/Lable on Bottle")

# Calculate relative depths within each group
df["Relative_Depth_From"] = groups.apply(lambda x: (x["Depth_From"] - x["Depth_From"].min()) / (x["Depth_To"].max() - x["Depth_From"].min()) * 100)
df["Relative_Depth_To"] = groups.apply(lambda x: (x["Depth_To"] - x["Depth_From"].min()) / (x["Depth_To"].max() - x["Depth_From"].min()) * 100)

# Save the updated data to a new Excel file
new_file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected_RD.xlsx"
df.to_excel(new_file_path, index=False)
```

```python
import pandas as pd
import os

# Read the Excel file
excel_file_path = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\CiASOM_SLF_Corrected.xlsx"
sheet_name = "All"
output_folder = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Device_Op_Num_files"

# Read the Excel file into a DataFrame
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)

# Get unique values in "PS Operation Device/Lable on Bottle" column
unique_devices = df["PS Operation Device/Lable on Bottle"].dropna().unique()

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Save separate CSV files for each device
for device in unique_devices:
    # Check if the value is a string
    if isinstance(device, str):
        # Replace forward slash ("/") in device name with underscore
        device_file_name = device.replace("/", "_")
        
        # Filter data for the current device
        filtered_data = df[df["PS Operation Device/Lable on Bottle"] == device]
        
        # Generate the CSV file path for the current device
        csv_file_path = os.path.join(output_folder, f"{device_file_name}.csv")
        
        # Save the filtered data as a CSV file
        filtered_data.to_csv(csv_file_path, index=False)
        
        print(f"CSV file saved for device: {device} - {csv_file_path}")
    else:
        print(f"Skipping device: {device} - Invalid value type")
```

```python
import os
import pandas as pd

input_folder = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Device_Op_Num_files"
output_folder = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Rel_Depth"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate over the CSV files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(input_folder, filename)
        
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Calculate the relative depth for each row
        mean_depth = (df["Depth_From"] + df["Depth_To"]) / 2
        max_depth = max(df["Depth_From"].max(), df["Depth_To"].max())
        relative_depth = ((mean_depth / max_depth) * 100).round(2)
        
        # Add the relative depth column to the DataFrame
        df["Rel_Depth_%"] = relative_depth
        
        # Save the updated DataFrame as a new CSV file
        output_file_path = os.path.join(output_folder, filename)
        df.to_csv(output_file_path, index=False)
        
        print(f"CSV file saved with relative depth: {output_file_path}")
```

```python
#### Heatmap Relative Humidity

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

input_folder = r"C:\CiASOM\PhD\07.Manuscripts\#2.Snow Surface\Data\Rel_Depth"

# Create a list to store all the dataframes
dfs = []

# Iterate over the CSV files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(input_folder, filename)
        
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Append the dataframe to the list
        dfs.append(df)

# Concatenate all dataframes into a single dataframe
merged_df = pd.concat(dfs)

# Convert the "Date" column to datetime type
merged_df["Date"] = pd.to_datetime(merged_df["Date"])

# Calculate the mean of "d18O" values for duplicate entries
merged_df = merged_df.groupby(["Date", "Rel_Depth_%"], as_index=False).mean()

# Set up the figure and axes for the heatmap
fig, ax = plt.subplots(figsize=(10, 8))

# Create the pivot table for the heatmap
pivot_table = merged_df.pivot(index="Rel_Depth_%", columns="Date", values="d18O")

# Generate the heatmap
sns.heatmap(pivot_table, cmap="coolwarm", cbar=True, ax=ax)

# Set the y-axis tick positions and labels
y_ticks = range(0, 101, 10)
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_ticks)

# Set the x-axis tick format
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))

# Rotate the x-axis tick labels for better visibility
plt.xticks(rotation=45)

# Set the axis labels and title
ax.set_xlabel("Date")
ax.set_ylabel("Relative Depth (%)")
ax.set_title("Time Series Heatmap of d18O by Relative Depth")

# Display the heatmap
plt.tight_layout()
plt.show()
```
